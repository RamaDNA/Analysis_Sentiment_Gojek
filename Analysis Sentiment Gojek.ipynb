## mencoba select data di tanggal 30 bulan 3

```python
new_df = df.reset_index(drop = True)
new_df.head()
```

```python
#add new field  tanggal
new_df['tanggal'] = new_df['at']

#change datetime to get new field
new_df['tanggal'] = pd.to_datetime(new_df['tanggal'])

#add month and day

new_df['time'] = new_df['tanggal'].dt.time
new_df['day'] = new_df['tanggal'].dt.day
new_df['month'] = new_df['tanggal'].dt.month
new_df['year'] = new_df['tanggal'].dt.year

new_df.head()
```

```python
#preprocessing text
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from  nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer #<<<<< cari semua fungsi import di samping

import re
```

```python
TEXT_CLEANING_RE = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"
stop_words = stopwords.words("indonesian")
stemmer = SnowballStemmer("indonesian") #TIDAK ada untuk varian bahasa indonesia <<<<<

```

```python
def preprocess(text, stem=False):
    text = re.sub(TEXT_CLEANING_RE,' ',str(text).lower()).strip()
    tokens = []
    for token in text.split():
        if token not in stop_words:
            if stem:
                tokens.append(stemmer.stem(token))
            else:
                tokens.append(token)
    return " ".join(tokens)
```

```python
new_df.content = new_df.content.apply(lambda x: preprocess(x))
```

```python
low_score = []
low_content = []
for x in new_df.index:
    if new_df.loc[x,'score'] <= 2:
        low_score.append(new_df.loc[x,'score'])
        low_content.append(new_df.loc[x,'content'])

print("finish")
```

```python
low_dataset_content = pd.DataFrame(low_score, columns=['score'])
low_dataset_content['text'] = low_content
```
